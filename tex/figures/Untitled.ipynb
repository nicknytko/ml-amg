```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_csv(fname):
    df = pd.read_csv(fname)
    v = df.values
    return v[:,1], v[:,2]
```

```python
it, loss = load_csv('learn_p.csv')
plt.figure(figsize=(10,4))
plt.plot(it, loss, label='Loss')
plt.plot(it, loss*0 + 0.6874, label='Baseline Lloyd')
plt.ylim((0.4, 1.0))
plt.xlabel('Training Iteration')
plt.ylabel('Approx conv. factor')
plt.grid()
plt.title('Loss function, learning $P$ directly.')
plt.legend()
plt.savefig('learn_p_directly.pdf')
```
